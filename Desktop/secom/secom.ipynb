```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("secom.data", delim_whitespace=True, header=None)
labels = pd.read_csv("secom_labels.data", delim_whitespace=True, header=None)
df['Target'] = labels[0]
df.head()
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(axis=1),inplace=True)
```

```python
df.shape
df.dtypes
```

```python
df.info()
```

```python
df.describe()
```

```python
dff=df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(dff)
plt.show()
```

```python
X=df.drop('Target',axis=1)
y=df['Target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Xts=scaler.fit_transform(X_train)
Xtests=scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(class_weight='balanced')
model.fit(Xts,y_train)
```

```python
# predict using the test features (Xtests), not the labels (y_test)
y_pred = model.predict(Xtests)
```

```python
np.unique(y_pred)
```

```python

plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual Values (y_test)")
plt.ylabel("Predicted Values (y_pred)")
plt.title("Predicted vs Actual Values")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_squared_error,accuracy_score,classification_report,confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc, RocCurveDisplay

print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
```
